```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
df_iphone = pd.read_csv("rssi-distance-iphone.csv")
df_iphone
```

```python
figure(figsize=(8, 6), dpi=80)

sns.lineplot(x=np.array(df_iphone.index), y='rssi_raw', data=df_iphone)
sns.lineplot(x=np.array(df_iphone.index), y='rssi_median', data=df_iphone)
```

```python
figure(figsize=(8, 6), dpi=80)
sns.lineplot(x=np.array(df_iphone.index), y='distance', data=df_iphone)
```
