```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
# Android phones (Pixel 4a, Samsung S10e)
id1 = 219873460
id2 = 390275417

# iPhones (XR, SE2)
id3 = 224577689
id4 = 251879309

# During the initial 5 minutes, all phones are next to each other, 
# then they are moved one by one to more than 3 meters away from original spot, 
# waiting 5 minutes after moving each phone, in this order:
# 219873460 (id1)
# 251879309 (id4)
# 390275417 (id2)
# 224577689 (id3)
# After 5 minutes with all phones next to each other again, two are moved together to a spot 
# 1 meter away from the other 2 and stay there for 5 more minutes. The ones moved are:
# 219873460 (id1)
# 251879309 (id4)
```

```python
df1 = pd.read_csv("run2/rssi-distance-android-219873460.csv")
df1.reset_index(inplace=True)
df1.timestamp = (df1.timestamp - df1.timestamp.min()) / 60.0
df12 = df1[df1['id'] == id2]
df13 = df1[df1['id'] == id3]
df14 = df1[df1['id'] == id4]
```

```python
figure(figsize=(15, 6), dpi=80)

df = df12
sns.lineplot(x='timestamp', y='rssi_raw', data=df)
sns.lineplot(x='timestamp', y='rssi_median', data=df)
sns.lineplot(x='timestamp', y='rssi_kalman', data=df)

figure(figsize=(15, 6), dpi=80)
sns.lineplot(x='timestamp', y='distance', data=df)
```
