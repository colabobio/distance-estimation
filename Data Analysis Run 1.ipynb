```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
id1 = 716634673
id2 = 690864300
id3 = 281371689
id4 = 327649634
```

```python
df1 = pd.read_csv("run1/rssi-distance-android-716634673.csv")
df1.reset_index(inplace=True)
df12 = df1[df1['id'] == id2]
df13 = df1[df1['id'] == id3]
df14 = df1[df1['id'] == id4]
```

```python
figure(figsize=(15, 6), dpi=80)

df = df12
sns.lineplot(x='index', y='rssi_raw', data=df)
sns.lineplot(x='index', y='rssi_median', data=df)
sns.lineplot(x='index', y='rssi_kalman', data=df)

figure(figsize=(15, 6), dpi=80)
sns.lineplot(x='index', y='distance', data=df)
```
